```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data["school_name"].count()
total_students = df["student_name"].count() 
total_budget = school_data["budget"].sum() 
average_math_score = df["math_score"].mean() 
average_reading_score = df["reading_score"].mean() 
passmath = df.loc[df["math_score"] > 69]
count_passmath= len(passmath)
percent_math = (count_passmath/total_students)*100
passreading = df.loc[df["reading_score"] > 69]
count_passread= len(passreading)
percent_read = (count_passread/total_students)*100
overall_pass = (percent_math + percent_read)/2
```

```python
summary_df = pd.DataFrame({"Total Schools": [total_schools],
                           "Total Students": [total_students],
                           "Total Budget": [total_budget],
                           "Average Math Score": [average_math_score],
                           "Average Reading Score": [average_reading_score],
                           "% Passing Math": [percent_math],
                           "% Passing Reading": [percent_read],
                           "% Overall Passing Rate": [overall_pass]})
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,}".format)
summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map("{:.2f}".format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map("{:.2f}".format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map("{:.2f}".format)
summary_df["% Passing Reading"] = summary_df["% Passing Reading"].map("{:.2f}".format)
summary_df["% Overall Passing Rate"] = summary_df["% Overall Passing Rate"].map("{:.2f}".format)

summary_df.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
df1 = school_data[["school_name","type","size","budget"]]# make a new data frame with selected columns
df1['Per Student Budget'] = df1['budget']/df1['size']
df2 = pd.DataFrame(df.groupby(["school_name"]).math_score.mean())

merge1 = pd.merge(df1, df2, how = "left", on=["school_name"])

df3 = pd.DataFrame(df.groupby(["school_name"]).reading_score.mean())

merge2 = pd.merge(merge1, df3, how = "left", on=["school_name"])

merge2 = merge2.rename(columns={"type":"School Type","size":"Total Students","budget":"Total Budget","math_score":"Average Math Score", "reading_score":"Average Reading Score"})

pass_maths = df.math_score > 69
maths = df[pass_maths]
df4 = pd.DataFrame(maths.groupby(["school_name"]).math_score.count())
df4 = df4.rename(columns={"math_score":"Number of students that pass math"})

merge3 = pd.merge(merge2, df4, how = "left", on=["school_name"])

pass_reads = df.reading_score > 69
reads = df[pass_reads]
df5 = pd.DataFrame(reads.groupby(["school_name"]).reading_score.count())
df5 = df5.rename(columns={"reading_score":"Number of students that pass reading"})

merge4 = pd.merge(merge3, df5, how = "left", on=["school_name"])

merge4['% Passing Math'] = (merge4['Number of students that pass math']/merge4['Total Students'])*100

merge4['% Passing Reading'] = (merge4['Number of students that pass reading']/merge4['Total Students'])*100

merge4['% Overall Passing Rate'] = (merge4['% Passing Math']+ merge4['% Passing Reading'])/2

overview = merge4.drop(columns="Number of students that pass math")
overview = overview.drop(columns="Number of students that pass reading")
overview.set_index('school_name', inplace=True)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
best_schools= overview.sort_values("% Overall Passing Rate", ascending=False)
best_schools["Total Budget"] = best_schools["Total Budget"].map("${:,}".format)
best_schools["Per Student Budget"] = best_schools["Per Student Budget"].map("${:.2f}".format)
best_schools["Total Students"] = best_schools["Total Students"].map("{:,}".format) 
best_schools["Average Math Score"] = best_schools["Average Math Score"].map("{:.2f}".format)
best_schools["Average Reading Score"] = best_schools["Average Reading Score"].map("{:.2f}".format)
best_schools["% Passing Math"] = best_schools["% Passing Math"].map("{:.2f}".format)
best_schools["% Passing Reading"] = best_schools["% Passing Reading"].map("{:.2f}".format)
best_schools["% Overall Passing Rate"] = best_schools["% Overall Passing Rate"].map("{:.2f}".format)

best_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
worst_schools= best_schools.sort_values("% Overall Passing Rate")

worst_schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grade9 = df.loc[df["grade"] == "9th", :]
grade9_math = pd.DataFrame(grade9.groupby(["school_name"]).math_score.mean())
grade9_math = grade9_math.rename(columns={"math_score":"9th"})

grade10 = df.loc[df["grade"] == "10th", :]
grade10_math = pd.DataFrame(grade10.groupby(["school_name"]).math_score.mean())
grade10_math = grade10_math.rename(columns={"math_score":"10th"})

grade11 = df.loc[df["grade"] == "11th", :]
grade11_math = pd.DataFrame(grade11.groupby(["school_name"]).math_score.mean())
grade11_math = grade11_math.rename(columns={"math_score":"11th"})

grade12 = df.loc[df["grade"] == "12th", :]
grade12_math = pd.DataFrame(grade12.groupby(["school_name"]).math_score.mean())
grade12_math = grade12_math.rename(columns={"math_score":"12th"})

math_grades1 = pd.merge(grade9_math, grade10_math, how = "left", on=["school_name"])
math_grades2 = pd.merge(math_grades1, grade11_math, how = "left", on=["school_name"])
math_grades = pd.merge(math_grades2, grade12_math, how = "left", on=["school_name"])
math_grades["9th"] = math_grades["9th"].map("{:.2f}".format)
math_grades["10th"] = math_grades["10th"].map("{:.2f}".format)
math_grades["11th"] = math_grades["11th"].map("{:.2f}".format)
math_grades["12th"] = math_grades["12th"].map("{:.2f}".format)

math_grades
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
grade9_reading = pd.DataFrame(grade9.groupby(["school_name"]).reading_score.mean())
grade9_reading = grade9_reading.rename(columns={"reading_score":"9th"})

grade10_reading = pd.DataFrame(grade10.groupby(["school_name"]).reading_score.mean())
grade10_reading = grade10_reading.rename(columns={"reading_score":"10th"})

grade11_reading = pd.DataFrame(grade11.groupby(["school_name"]).reading_score.mean())
grade11_reading = grade11_reading.rename(columns={"reading_score":"11th"})

grade12_reading = pd.DataFrame(grade12.groupby(["school_name"]).reading_score.mean())
grade12_reading = grade12_reading.rename(columns={"reading_score":"12th"})

reading_grades1 = pd.merge(grade9_reading, grade10_reading, how = "left", on=["school_name"])
reading_grades2 = pd.merge(reading_grades1, grade11_reading, how = "left", on=["school_name"])
reading_grades = pd.merge(reading_grades2, grade12_reading, how = "left", on=["school_name"])

reading_grades["9th"] = reading_grades["9th"].map("{:.2f}".format)
reading_grades["10th"] = reading_grades["10th"].map("{:.2f}".format)
reading_grades["11th"] = reading_grades["11th"].map("{:.2f}".format)
reading_grades["12th"] = reading_grades["12th"].map("{:.2f}".format)

reading_grades
```
